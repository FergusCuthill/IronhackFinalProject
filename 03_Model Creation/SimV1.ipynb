```python
import numpy as np
import pandas as pd
from scipy import sparse
```

```python
# Import user games, steam game data and other user data
myg = pd.read_excel("../01_Raw Data/Data Files/mygames.xlsx")
sg = pd.read_csv("../02_Data Cleaning/Cleaned Data/clean_sg.csv")
```

```python
# This section saved the csv matrix to a sparse matrix format improving processing times. No need to run this again. 
# ud = pd.read_csv("../02_Data Cleaning/Cleaned Data/150k_Users.csv", skiprows=[0,2]).drop("appid", axis=1)
# ug = sparse.coo_matrix(ud.astype("int32").values).tocsr()
# sparse.save_npz("Data/User_Data.npz", ug)

# Load from sparse file. Still need column headers from csv
ug = sparse.load_npz("Data/User_Data.npz")
mygames = pd.read_csv("../02_Data Cleaning/Cleaned Data/150k_Users.csv", skiprows=[0,2], nrows=0).drop("appid", axis=1).T
```

```python
# Create dataframe with user games in
mygames["mg"] = [(int(col) in list(myg.appid))*1 for col in mygames.index]
# Convert dataframe to sparse matrix
mg = sparse.coo_matrix(mygames.astype("int32").values).tocsr()
# Dot product of matricies to get most similar users. 
out = ug.dot(mg).sum(1)
# Extract indicies of 10 best matches. 
tt = pd.DataFrame(out).nlargest(10, 0).sort_index()
# Weight scores in matrix by similarity and then sum to get total score by game
s = ug[tt.index].multiply(np.array([[i] for i in tt[0]])).sum(0)
mygames["scores"] = s.tolist()[0]
```
