```python
import pandas as pd
import numpy as np
pd.set_option("max_columns", 100)
```

```python
ssal = pd.read_csv('../01_Raw Data/Data Files/steam_app_list.csv')
sad = pd.read_csv('../01_Raw Data/Data Files/steam_app_data.csv')
sspd = pd.read_csv('../01_Raw Data/Data Files/steamspy_data.csv')
```

```python
# Firstly select the columns from the steam dataframe which may be useful to me. Also check against steamspy columns to see if data is in a better format there. 
sadcols = ['type', 'name', 'steam_appid', 'required_age', 'is_free', 'controller_support', 'dlc', 'pc_requirements', 'developers', 
           'publishers', 'price_overview', 'platforms', 'metacritic', 'categories', 'recommendations', 'achievements', 'release_date']

sad = sad[sadcols]
```

```python
# Select the additional columns from steamspy that might be useful. Attempt to avoid duplicate info. 
sspdcols = ['appid', 'score_rank', 'positive', 'negative', 'userscore', 'owners', 'average_forever', 'average_2weeks', 'median_forever', 
            'median_2weeks', 'price', 'initialprice', 'discount', 'languages', 'genre', 'ccu', 'tags']

sspd = sspd[sspdcols]
```

```python
# Join sad to sspd
sg = sad.set_index("steam_appid").join(sspd.set_index("appid"), how="inner", lsuffix="st", rsuffix="sp").reset_index().rename({"index":"appid"}, axis=1)
sg.head()
```

```python
sg.to_csv('Cleaned Data/stgames.csv', index=False)
```
